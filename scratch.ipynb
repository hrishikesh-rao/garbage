```python
import matplotlib.pyplot as plt

# Sample data
x_values = [1, 2, 3, 4, 5]
y_values = [i**2 for i in x_values]  # squares

# Create line plot
plt.plot(x_values, y_values, marker="o")
plt.title("Simple Line Plot: Squares of Numbers")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 100)

# Compute sine of x
y = np.sin(x)

# Plot the sine graph
plt.plot(x, y, label='sin(x)', color='blue')
plt.title("Sine Wave")
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create x values from 0 to 20 with step 0.1
x = np.arange(0, 20., 0.0001)

# Compute sine of x
y = np.sin(x)

# Plot the sine graph
plt.plot(x, y, label='sin(x)', color='purple')
plt.title("Sine Wave from 0 to 20")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.legend()
plt.show()
```
